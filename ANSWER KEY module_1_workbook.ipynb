# Your Turn With a Shopping List

**Given the shopping list below, perform the following tasks:**


* Append "apples" to the shopping list & print the new list.
* Insert "orange juice" at the second position & print the new list.
* Change "bread" to "whole grain bread" & print the new list.
* Print the first and last items from the shopping list.
* Print a slice of the shopping list that includes the second and third items.


```python
shopping_list = ["milk", "eggs", "bread", "butter"]

# 1. Append "apples" to the shopping list
shopping_list.append("apples")
print(shopping_list)

# 2. Insert "orange juice" at the second position
shopping_list.insert(1, "orange juice")
print(shopping_list)

# 3. Change "bread" to "whole grain bread"
shopping_list[3] = "whole grain bread"
print(shopping_list)

# 4. Print the first and last items from the shopping list
print("First item:", shopping_list[0])
print("Last item:", shopping_list[-1])

# 5. Print a slice of the shopping list that includes the second and third items
print("Slice:", shopping_list[1:3])
```

```python
#Your code here
# Indexing and slicing a list
fruits = ["apple", "banana", "cherry", "orange", "blueberry"]

# Use string concatenation with elements in the list to display something new
new_sentence = fruits[1] + " " + fruits[3]

# Obtain user input
user_input = input("Enter a number between 0 and 4: ")
index = int(user_input)

# Use an F-string
print(f"The fruit at index {index} is {fruits[index]}.")

# Print a slice of the list
print(f"The fruits from index 1 to 3 are: {fruits[1:4]}")

# Print the new sentence
print("New sentence:", new_sentence)
```
